```python
COMPLETIONS_MODEL = "text-demo"

prompt = f"""
You are a pirate and answers questions.
Q: who is the captain?
A: """


```

```python
response = client.chat.completions.create(
    model= COMPLETIONS_MODEL,
    messages = [{"role" : "system", "content" : "You are a AI tutor"},
                {"role" : "user", "content" : "What is AI"}],

    temperature = 0.7,
    max_tokens = 800,
    top_p = 0.95,
    frequency_penalty= 0,
    presence_penalty= 0,
    stop = None)


#print(response)

print(response.choices[0].message.content)
```

```python
response = client.chat.completions.create(
    model= COMPLETIONS_MODEL,
    messages = [{"role" : "system", "content" : "You are a Cake Baker"},
                {"role" : "user", "content" : "Get me a recipe to make a cake"}],
    n=3
    )

for choice in response.choices :
  print(choice.message.content)
```

```python
import pandas as pd

df_orig = pd.read_csv("bbc-news-data.csv", delimiter='\t')

df = df_orig.copy()

df
```

```python
df.describe()
```

```python
for col in ['category', 'filename'] :
  print(df[col].unique())
```

```python
df['category'].value_counts().plot(kind='bar',
                                    xlabel = 'Category',
                                    ylabel = 'Articles Count',
                                    rot = 0,
                                    title = 'Number of articles per Category')
```

```python
df["word_count"] = df["content"].apply(lambda n : len(n.split()))
df
```

```python
import seaborn as sns

sns.set_style("whitegrid")
sns.catplot(data=df, x="category", y = "word_count", kind = "box", height=5)
```
